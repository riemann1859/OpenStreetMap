```python
import xml.etree.cElementTree as ET
add='D:/Anaconda3/mongodb-json-files/chicago.osm'
tree=ET.ElementTree(file=add)
root=tree.getroot()
```

```python
import re

#check whether keys contain problem characters or not
#if it does, ignore it
problemchars = re.compile(r'[=\+/&<>;\'"\?%#$@\,\.\t\r\n]')




for elem  in root: 
#run through only top level element
    if elem.tag=='node' or elem.tag=='way':
        for sub_elem in elem:
            if sub_elem.tag=='tag':
                if problemchars.search(sub_elem.attrib['k']):
                    print(elem.tag, elem.attrib)
                    for item in elem:
                        print(item.tag,item.attrib)
```

```python
# by the above query, we know that there is  No key with problem characters



CREATED=[ "version", "changeset", "timestamp", "user", "uid"]

ADDRESS=[]

                    
                    
for elem  in root: 
    if elem.tag=='node' or elem.tag=='way':
        for sub_elem in elem:
            if sub_elem.tag=='tag':
                if sub_elem.attrib['k'].strip().lower().startswith("add"):
                    ADDRESS.append(sub_elem.attrib['k'])
                
                
set(ADDRESS)
```

```python
#it's expected that addr:full=addr:housenumber+addr:street
#are there any exceptions to this?
#exceptions for addr:full

for elem  in root: 
    if elem.tag=='node' or elem.tag=='way':
        full=""
        street=""
        housenumber=""
        for sub_elem in elem:
            if sub_elem.tag=='tag':
                if sub_elem.attrib['k']=='addr:full':
                    full=sub_elem.attrib['v']
                elif sub_elem.attrib['k']=='addr:housenumber':
                    housenumber=sub_elem.attrib['v']
                else:
                    if sub_elem.attrib['k']=='addr:street':
                        street=sub_elem.attrib['v']
        if full!="" and full!=housenumber+" "+street:
            print(elem.tag,elem.attrib)
            for item in elem:
                print(item.tag, item.attrib)
```

```python
#remove the sub-elements with key addr:full
#deal with explicitly some exceptions for addr:full

#'k': 'addr:full', 'v': '50 S Fairbank St, Addison, IL 60101'
#'k': 'addr:full', 'v': '1314 W. Wrightwood Chicago, IL'
#'k': 'addr:full', 'v': '8361 W Belmont, River Grove, IL 60171'



attrib_dict11={'k':'addr:housenumber', 'v':'50'}
attrib_dict12={'k':'addr:city', 'v':'Addison'}
attrib_dict13={'k':'addr:state', 'v':'IL'}
attrib_dict14={'k':'addr:street', 'v':'S Fairbank St'}
attrib_dict15={'k':'addr:postcode', 'v':'60101'}

attrib_dict21={'k':'addr:housenumber', 'v':'1314'}
attrib_dict22={'k':'addr:city', 'v':'Chicago'}
attrib_dict23={'k':'addr:state', 'v':'IL'}
attrib_dict24={'k':'addr:street', 'v':'W. Wrightwood Avenue'}




for elem  in root: 
    if elem.tag=='node' or elem.tag=='way':
        if  elem.tag=='node' and elem.attrib['id']=='346207271':
            #to add child element we can also use append method: elem.append
            ET.SubElement(elem,'tag',attrib=attrib_dict21)
            ET.SubElement(elem, 'tag',attrib=attrib_dict22)
            ET.SubElement(elem, 'tag', attrib=attrib_dict23)
            ET.SubElement(elem, 'tag',attrib=attrib_dict24)
                 
            for sub_elem in elem:
                if sub_elem.tag=='tag':
                    if sub_elem.attrib['k']=='addr:full':
                    # remove sub_elem from the parent element elem
                        elem.remove(sub_elem)  
                    
        elif  elem.tag=='node' and elem.attrib['id']=='1117871207':
            
            for sub_elem in elem:
                if sub_elem.tag=='tag':
                    if sub_elem.attrib['k']=='addr:full':
                        elem.remove(sub_elem)  
                    if sub_elem.attrib['k']=='addr:street':
                        sub_elem.attrib['v']='West Belmont Avenue'
        
        elif elem.tag=='node' and  elem.attrib['id']=='2111412795':
            
            ET.SubElement(elem,'tag',attrib=attrib_dict11)
            ET.SubElement(elem, 'tag',attrib=attrib_dict12)
            ET.SubElement(elem, 'tag', attrib=attrib_dict13)
            ET.SubElement(elem, 'tag',attrib=attrib_dict14)
            ET.SubElement(elem, 'tag',attrib=attrib_dict15)
                 
            for sub_elem in elem:
                if sub_elem.tag=='tag':
                    if sub_elem.attrib['k']=='addr:full':
                        elem.remove(sub_elem)  
        
        else:
            for sub_elem in elem:
                if sub_elem.tag=='tag':
                    if sub_elem.attrib['k']=='addr:full':
                        elem.remove(sub_elem)  
            
            
        
        
        
        
tree.write("chicago.osm",encoding='UTF-8')  
                    
```

```python
#convert addr to addr:street
#convert add:city to addr:city

for elem in root:
    if elem.tag=='node' or elem.tag=='way':
        for sub_elem in elem:
            if sub_elem.tag=='tag':
                if sub_elem.attrib['k']=='addr':
                    sub_elem.attrib['k']='addr:street'
                if sub_elem.attrib['k']=='add:city':
                    sub_elem.attrib['k']='addr:city'
                    
tree.write("chicago.osm",encoding='UTF-8')  
```

```python
#check street names contain the characters: ,:-#


chars_=re.compile(r'[,#\-:]')

for elem in root:
    if elem.tag=='node' or elem.tag=='way':
        for sub_elem in elem:
            if sub_elem.tag=='tag':
                if sub_elem.attrib['k']=="addr:street":
                    if len(chars_.findall(sub_elem.attrib['v']))>0:
                        print(chars_.findall(sub_elem.attrib['v']))
                        print(elem.tag, elem.attrib)
                        for item in elem:
                            print(item.tag, item.attrib) 
```

```python
#   U.S. Routes or U.S. Highways
# Two type of naming: US Route 59 or Highway 59
# use state code to represent the part of the route running throughout the state: e.g. US Route IL 59

# US-6 ----->  US Route 6

for elem  in root: 
    if elem.tag=='node' or 'way':
        for sub_elem in elem:
            if sub_elem.tag=='tag':
                if sub_elem.attrib['k']=='addr:street':
                    if sub_elem.attrib['v']=='US-6':
                        sub_elem.attrib['v']='US Route 6'
                    
    
        
        
        
tree.write("chicago.osm",encoding='UTF-8')  
```
